# Pandas Visulaization

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv(r"D:\Data Analyst Bootcamp\Pandas for Data Analytics\data\Ice Cream Ratings.csv")
df = df.set_index('Date')
```

```python
df.plot(kind='line')
```

```python
df.plot(kind='line', subplots=True)
```

```python
df.plot(kind='line', title="Ice Cream Ratings", xlabel="Daily Rating", ylabel='Scores')
```

```python
df.plot(kind='bar')
```

```python
df.plot(kind='bar', stacked=True)
```

```python
df.plot.barh(stacked=True)
```

```python
df.plot.scatter(x = 'Texture Rating', y = 'Overall Rating', s = 300, c = 'Orange')
```

```python
df.plot.hist(bins = 10)
```

```python
df.boxplot()
```

```python
df.plot.area(figsize=(10,5))
```

```python
df.plot.pie(y='Flavor Rating', figsize = (10,10))
```
